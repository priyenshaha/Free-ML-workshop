```python
import pandas as pd
from scipy import stats as st

#pandas data stream
data=pd.read_csv('.\datasets\Apparel_Data.csv')
data
```

```python
#pandas data series
usage=data.Usage
gender=data.Gender
awareness=data.Awareness
pref=data.Preference
```

```python
genUse=pd.crosstab(gender,usage)
genUse
```

```python
genPref=pd.crosstab(gender,pref)
genPref
```

```python
genAware=pd.crosstab(gender,awareness)
genAware
```

```python
#used to find P-value (2nd arg) to analyze dependency
#if P-value is < 0.05, the variables are dependent
st.chi2_contingency(genUse)
```

```python
t2 = st.chi2_contingency(genAware)
t2[1]
```

```python
t3 = st.chi2_contingency(genPref)
t3
```
